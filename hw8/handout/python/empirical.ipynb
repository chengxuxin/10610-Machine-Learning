```python
from q_learning import *
import matplotlib.pyplot as plt
from collections import deque
```

```python
def get_rollingmean(returns):
    rollingmean = deque(maxlen=25)
    mean_return = []
    for r in returns:
        rollingmean.append(r)
        mean_return.append(np.mean(rollingmean))
    return mean_return
```

```python
mode = "raw"
episodes = 2000
max_iterations = 200
epsilon = 0.05
gamma = 0.999
learning_rate = 0.001

env = MountainCar(mode=mode)
agent = QLearningAgent(env, mode=mode, gamma=gamma, epsilon=epsilon, lr=learning_rate)
returns = agent.train(episodes, max_iterations)
```

```python
mean_return = get_rollingmean(returns)
plt.figure(dpi=200)
plt.plot(returns)
plt.plot(mean_return)
plt.xlabel("episode")
plt.ylabel("episode return")
```
